```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'..\data\preludium_with_additional_info.csv')
df
```

```python
plt.figure(figsize=(15,8))
hs = ["#E12647" for x  in range(6)]
nz = ["#798897" for x  in range(9)]
st = ["#E29897" for x  in range(10)]
allcls = hs+nz+st
ho = ['HS'+str(i+1) for i in range(6)]
nz = ['NZ'+str(i+1) for i in range(9)]
st = ['ST'+str(i+1) for i in range(10)]
order =ho+nz+st 
sns.barplot(x='subpanel', y='budget',data=df,palette=allcls,order=order )
plt.xlabel('Panel')
plt.ylabel('Średni budżet')
sns.despine(top=True, right=True, left=True, bottom=True)
#plt.title("Średni budżet na panel", fontdict=None, loc='center', pad=None)
plt.savefig('bar_panels_budgets.png', format='png', bbox_inches='tight')
#25
```

```python
panel_df = df.copy()
panel_df["subpanel"] =panel_df["subpanel"].str.replace('\d+', '')
panel_df
```

```python
plt.figure(figsize=(7,8))
sns.barplot(x='subpanel', y='budget',data=panel_df )
```

```python
panel_with_budget = panel_df.groupby(['subpanel'])['budget'].sum().reset_index(name='budget')
panel_with_budget
```

```python
hs_budget=panel_with_budget[panel_with_budget.subpanel == 'HS'].budget.values[0]
nz_budget=panel_with_budget[panel_with_budget.subpanel == 'NZ'].budget.values[0]
st_budget=panel_with_budget[panel_with_budget.subpanel == 'ST'].budget.values[0]
sum_budget = hs_budget+nz_budget+st_budget
hs_budget+nz_budget+st_budget
```

```python

# Pie chart
labels = ['Nauki Humanistyczne - HS', 'Nauki o Ziemi - NZ', 'Nauki ścisłe i Techniczne - ST']
sizes = [hs_budget/sum_budget*100, nz_budget/sum_budget*100, st_budget/sum_budget*100]
explode = (0, 0.1, 0)  
fig1, ax1 = plt.subplots()
#plt.figure(figsize=(5.80, 4.00), dpi=100)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90,colors=['#E12647','#E29897','#798897'])
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
#plt.title("Procent pieniędzy jaki przeznaczono przez wszystkie lata \n na poszczególne kategorie", fontdict=None, loc='center', pad=None)
plt.savefig('pie_panels_budgets.png', format='png', bbox_inches='tight',Transparent=True)
```

```python
desc = pd.read_csv(r'..\data\preludium_descriptors.csv')
desc_df = df.merge(desc, left_on = 'id', right_on = 'Id')
desc_df
```

```python
mean = desc_df.groupby(['Descriptors']).agg({'budget':['mean','std']})
mean.columns = ['mean','std']
mean.reindex(columns=sorted(mean.columns))
mean = mean.sort_values(by=['mean'], ascending=False)
mean = mean.head(10)
mean = mean.reset_index()
data = desc_df[desc_df['Descriptors'].isin(mean['Descriptors'])]
data = data.sort_values(by=['budget'], ascending=False)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x='Descriptors', y='mean',data=mean )
plt.xlabel('Subpanel')
plt.ylabel('Średni budżet')
plt.title("Średni budżet na subpanel", fontdict=None, loc='center', pad=None)
plt.savefig('subpanels_budgets.svg', format='svg', bbox_inches='tight')
```
