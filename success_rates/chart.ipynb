```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = pd.read_csv('../data/success_rates.csv')
df['id'] = pd.to_numeric(df['name'].str[9:])
df = df.drop(columns=['name'])
df = df.set_index('id')
df.head()
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(8,4)
g = sns.lineplot(data=df, ax=ax, linewidth=2)
ax.set_xlim(1, len(df))
ax.set_xticks(range(1, len(df)+1))
ax.set(xlabel='edycja PRELUDIUM')
plt.legend(title='Wnioski', labels=['Złożone', 'Zakwalifikowane'])
plt.show()
```

```python
df['percent'] = df['qualified']/df['filled']
fig, ax = plt.subplots() 
fig.set_size_inches(8,4)
g = sns.lineplot(x='id', y='percent', data=df.reset_index(), ax=ax, linewidth=2)
ax.set_xlim(1, len(df))
ax.set_xticks(range(1, len(df)+1))
ax.set(xlabel='edycja PRELUDIUM', ylabel='procent wniosków zakwalifikowanych')
plt.show()
```
